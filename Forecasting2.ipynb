```python
import numpy as np
import pandas as pd
```

# Simple exponential smoothening

```python
#read data
data=pd.read_csv("data1.csv")
```

```python
#Initialising arrays and first values
X=data['Data']
alpha=0.1
Forecast=np.ones((len(data)))
Forecast[0]=X[0]

#update values
for i in range(1,len(data)):
    Forecast[i]=alpha*X[i]+(1-alpha)*Forecast[i-1]
data['Forecast']=pd.DataFrame(Forecast)
data
```

# Holt's method (trend is present)

```python
#Initialising arrays and first values
alpha=0.9
beta=0.9
level=np.ones((len(data)))
trend=np.ones((len(data)))
forecast2=np.ones((len(data)))

level[0]=X[0]
trend[0]=X[1]-X[0]
forecast2[0]=level[0]+trend[0]

for i in range(1,len(data)):
    level[i]=alpha*X[i]+(1-alpha)*(level[i-1]+trend[i-1])
    trend[i]=beta*(level[i]-level[i-1])+(1-beta)*trend[i-1]
    forecast2[i]=level[i]+trend[i]
#creating a dataframe and showing values      
data2=pd.DataFrame()
data2['data']=X    
data2['Level']=pd.DataFrame(level)
data2['Trend']=pd.DataFrame(trend)
data2['Forecast']=pd.DataFrame(forecast2)
data2
```

# Holt-Winterâ€™s Method (When Trend and Seasonality are Present)

```python
# Read data
X=pd.read_csv('data2.csv')['Data']

#Initialise variables
alpha=0.716
beta=0
gamma=0.993
season_len=12
len_data=len(X)
num_seasons=len_data/season_len

#Initialising arrays and first values
smooth=np.ones((len(X)))
trend=np.ones((len(X)))
season_avg=np.ones((len(X)))
for i in range(int(num_seasons)):
    season_avg[12*i:12*(i+1)]=sum(X[12*i:12*(i+1)])/12
deviation=X-season_avg    
seasonal=np.ones((len(X)))
result=np.ones((len(X)))

smooth[0]=result[0]=X[0]
trend[0]=(sum(X[season_len:season_len*2])-sum(X[0:season_len]))/season_len**2

for i in range(12):
    seasonal[i]=sum(deviation[i::season_len])/num_seasons
#updating values    
for i in range(1,len(X)):
    smooth[i]=alpha*(X[i]-seasonal[i])+(1-alpha)*(smooth[i-1]+trend[i-1])
    trend[i]=beta*(smooth[i]-smooth[i-1])+(1-beta)*trend[i-1]
    if (i+season_len)< len(X):
        seasonal[i+season_len]=gamma*(X[i]-smooth[i])+(1-gamma)*seasonal[i]
    result[i]=smooth[i]+trend[i]+seasonal[i]

#creating a dataframe and showing values    
data3=pd.DataFrame()
data3['data']=X    
data3['Smooth']=pd.DataFrame(smooth)
data3['Trend']=pd.DataFrame(trend)
data3['Seasonal']=pd.DataFrame(seasonal)
data3['Result']=pd.DataFrame(result)
data3.head()
```
